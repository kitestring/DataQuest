```python
%matplotlib inline
import pandas as pd
import matplotlib

# *******************
# Read the dataset into a DataFrame and start exploring the data

recent_grads = pd.read_csv('recent-grads.csv')
first_row = recent_grads.iloc[1]
print("**********\nFirst Row of Table")
print(first_row)

print("**********\nHead - 5")
print(recent_grads.head(5))

print("**********\nTail - 5")
print(recent_grads.tail(5))

print("**********\nTable Description")
print(recent_grads.describe())

print("**********\nTable Shape")
raw_data_count = recent_grads.shape[0]
print("Rows: ", raw_data_count)
print("Columns: ", recent_grads.shape[1])

# Data Cleaning......
# Drop rows with missing values. Matplotlib expects that 
# columns of values we pass in have matching lengths and 
# missing values will cause matplotlib to throw errors.
recent_grads.dropna(inplace=True)
cleaned_raw_data_count = recent_grads.shape[0]

print("**********\nData Cleaning")
print("Raw Data - Rows: ", raw_data_count)
print("Cleaned Data - Rows: ", cleaned_raw_data_count)
```
